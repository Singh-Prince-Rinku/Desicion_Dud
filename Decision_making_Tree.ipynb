```python
import pandas as pd
```

```python
dataset = pd.read_csv('Decision.csv')
```

```python
dataset
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
outlook = LabelEncoder()
humidity=LabelEncoder()
windly=LabelEncoder()
play=LabelEncoder()
```

```python
dataset['outlook']=outlook.fit_transform(dataset['outlook'])
dataset['humidity']=outlook.fit_transform(dataset['humidity'])
dataset['windly']=outlook.fit_transform(dataset['windly'])
dataset['play']=outlook.fit_transform(dataset['play'])
```

```python
dataset
```

```python
features_cols = ['outlook','humidity','windly']
x = dataset[features_cols]
y = dataset.play
```

```python
print(x)     
```

```python
y
```

```python
from sklearn.model_selection import  train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.tree import  DecisionTreeClassifier
```

```python
classifier=DecisionTreeClassifier(criterion='entropy')
```

```python
classifier.fit(X_train,y_train)
```

```python
classifier.predict(X_test)
```

```python
X_test
```

```python
classifier.score(X_test,y_test)
```

```python
from sklearn import tree
tree.plot_tree(classifier)
```
